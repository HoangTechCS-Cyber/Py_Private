# Đọc và hiển thị hình ảnh

```python
import cv2 as cv
img = cv.imread("/content/Avatar.jpg")

#Opencv
#cv.imshow("Avatar", img)
#cv.waitKey(0)
# Matplotib
import matplotlib.pyplot as plt

#chuyen doi tu opencv sang maplotlib
img_plt = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plt)
#
plt.title("Avatar")
plt.axis("on")
plt.show()

    # Lấy kích thước ảnh
height, width, channels = img.shape
print(f"Kích thước ảnh gốc: {width} x {height}")
print(f"Số kênh màu: {channels}")
```

# 2.Xử lý ảnh cơ bản

## 2.1.Thay đổi kích thước và cắt ảnh
  - Mục tiêu
    - Resize ảnh
    - Cắt ảnh
    

```python

# Thay đổi kích thước ảnh
resize_img = cv.resize(img, (960, 1280))  # Resize ảnh thành 960x1280 pixel

# Cắt ảnh
crop_img = img[50:200, 100:300]  # Cắt từ tọa độ (100, 50) đến (300, 200)

# Chuyển đổi ảnh sang RGB để hiển thị đúng màu với Matplotlib
resize_img_rgb = cv.cvtColor(resize_img, cv.COLOR_BGR2RGB)
crop_img_rgb = cv.cvtColor(crop_img, cv.COLOR_BGR2RGB)

# Hiển thị ảnh resize
plt.imshow(resize_img_rgb)
plt.title("Resized Image")
plt.axis("off")
plt.show()

# Hiển thị ảnh crop
plt.imshow(crop_img_rgb)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```
